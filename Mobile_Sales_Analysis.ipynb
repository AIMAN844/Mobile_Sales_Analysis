# Exploratory Data Analysis

**Checking Duplicates records**

```python
df_sales_data.duplicated()
```

```python
df_sales_data['date'] = pd.to_datetime(df_sales_data['Dispatch Date'])
```

```python
# Convert 'Dispatch Date' to datetime
df_sales_data['Dispatch Date'] = pd.to_datetime(df_sales_data['Dispatch Date'], errors='coerce')

# Now extract the year
df_sales_data['year'] = df_sales_data['Dispatch Date'].dt.year
```

```python
df_sales_data.head(2)
```

```python
# Set general plot style
sns.set(style='whitegrid')
plt.figure(figsize=(12, 6))
```

```python
# 1. Count plot of Product Types
plt.subplot(2,2,1)
sns.countplot(data=df_sales_data, x= 'Product')
plt.title('Distribution of Product Types')
plt.xticks(rotation=45)
plt.show()
```

KT: Distribution of Product Types: The dataset includes both laptops and mobile phones, with mobile phones appearing more frequently.

```python
# 2. Count plot of Brand Types
plt.figure(figsize=(12, 6))
plt.subplot(2,2,1)
sns.countplot(data=df_sales_data, x= 'Brand')
plt.title('Distribution of Brand Types')
plt.xticks(rotation=90)
plt.show()
```

```python
# 3. Top 10 Brands by Quantity Sold
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 2)
top_brands = df_sales_data.groupby('Brand')['Quantity Sold'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_brands.values, y=top_brands.index)
plt.title("Top 10 Brands by Quantity Sold")
plt.show()
```

KT: Top 10 Brands by Quantity Sold: Certain brands dominate in sales volume; this can inform which brands are most popular or well-stocked.

```python
# 4. Distribution of Prices
plt.figure(figsize=(18, 6))
plt.subplot(2,2,2)
sns.histplot(df_sales_data['Price'],bins=30, kde=True)
plt.title("Distribution of Product Prices")
plt.show()
```

KT: Price Distribution: Product prices are right-skewedâ€”most products are moderately priced, with a few premium-priced ones.

```python
# 5. Average Price by Product Type
plt.figure(figsize=(12, 6))
plt.subplot(2,2,4)
sns.boxplot(data=df_sales_data, x = 'Product', y='Price')
plt.title("Price Distribution by Product Type")
plt.tight_layout()
plt.show()
```

KT: Price by Product Type: Laptops generally have a broader and higher price range than mobile phones.

```python
plt.figure(figsize=(16, 12))

```

```python
# 6. Quantity Sold over Regions
plt.subplot(3,2,1)
region_sales = df_sales_data.groupby('Region')['Quantity Sold'].sum().sort_values(ascending=False)
sns.barplot(x=region_sales.values, y= region_sales.index)
plt.title("Total Quantity Sold by Region")
plt.show()
```

KT: Quantity Sold by Region: Some regions, such as Central and South, contribute more significantly to overall sales.

```python
# 7 Average Quantity Sold per Brand (Top 10)
plt.figure(figsize=(12, 6))
plt.subplot(3,2,2)
avg_brand_sales = df_sales_data.groupby('Brand')['Quantity Sold'].mean().sort_values(ascending=False).head(10)
sns.barplot(x = avg_brand_sales.values, y = avg_brand_sales.index)
plt.title("Top 10 Brands by Avg Quantity Sold per Order")
plt.show()
```

KT : Top Brands by Average Quantity Sold per Order: Some brands may have fewer transactions but higher quantities sold per orderâ€”suggesting bulk or B2B sales.

```python
# . Monthly Sales Trend
plt.figure(figsize=(12, 6))
df_sales_data['Dispatch Date'] = pd.to_datetime(df_sales_data['Dispatch Date'])
df_sales_data['Month'] = df_sales_data['Dispatch Date'].dt.to_period('M')
monthly_sales = df_sales_data.groupby('Month')['Quantity Sold'].sum().reset_index()
monthly_sales['Month'] = monthly_sales['Month'].astype(str)
plt.subplot(3,2,3)
sns.lineplot(data=monthly_sales, x='Month', y='Quantity Sold', marker='o')
plt.title("Monthly Quantity Sold Trend")
plt.xticks(rotation=90)
plt.show()
```

KT: Monthly Sales Trend: There's noticeable fluctuation in sales across months, which could indicate seasonal demand or promotional events.

```python
# 9. Heatmap: Correlation among numeric features
plt.figure(figsize=(12, 6))
plt.subplot(3, 2, 6)
corr = df_sales_data[['Price', 'Quantity Sold']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")

plt.tight_layout()
plt.show()
```

KT: Correlation Heatmap: Price and Quantity Sold have a weak or no linear correlation, implying other factors may drive quantity sold.

```python
# 10. ROM vs Price
plt.figure(figsize=(12, 6))
plt.subplot(3, 2, 4)
sns.boxplot(data=df_sales_data, x='ROM', y='Price')
plt.title("Price Distribution by ROM Size")
plt.xticks(rotation=45)
plt.show()
```

KT: Price by ROM Size: Similarly, higher ROM (like 512GB) is associated with premium pricing.
